###Ejemplo de if

```python
#1. Ejemplo de if simple (con operadores relacionales)
#Se utiliza para ejecutar un bloque de código solo si se cumple una condición específica
edad = 20
if edad >= 18:
    print("Eres mayor de edad.")
```

```python
#2. Ejemplo de if-else (con operadores aritméticos y relacionales)
#Permite manejar dos alternativas: una si la condición es verdadera y otra si es falsa

numero = 10
if numero % 2 == 0:
    print("El número es par.")
else:
    print("El número es impar.")
```

```python
#3. Ejemplo de if-elif-else (con operadores lógicos y relacionales)
#Se usa cuando existen más de dos posibilidades o rangos que evaluar
puntuacion = 100
if puntuacion >= 90:
    print("Calificación: A")
elif puntuacion >= 80 and puntuacion < 90:
    print("Calificación: B")
else:
    print("Calificación: C o menor")
```
